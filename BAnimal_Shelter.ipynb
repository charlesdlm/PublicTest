```sql
-- natural keys he uses alias
-- join using name and species which is the business id for the animal 
-- we start with the from condition and then add in the select

SELECT	AD.*, A.Breed, A.Implant_Chip_ID
FROM	Animals AS A
		INNER JOIN 
		Adoptions AS AD
		ON	AD.Name = A.Name 
			AND 
			AD.Species = A.Species;
```

```sql
-- for inner joins the order of the joins does not matter 
-- this returns 70 records 

SELECT	*
FROM	Animals AS AN
		INNER JOIN 
		Adoptions AS AD
			ON AD.Name = AN.Name 
			AND 
			AD.Species = AN.Species
		INNER JOIN 
		Persons AS P 
			ON	AD.Adopter_Email = P.Email;
```

```sql
-- however when joining using inner and outer joins then you need to be careful. This next query still gives 70 rows, although you 
-- think that it was going to be 100 (the total rows in the animal table). Thats because the inner join is last. 
-- Here first animals is joined with adoptions, to create a data set off 100 rows and then there is an inner join of using the email
-- address to limit it to 70 rows. Ie 3 tables, 2 joins. The outer join should be the last one...see next step

SELECT	*
FROM	Animals AS AN
		LEFT OUTER JOIN 
		Adoptions AS AD
			ON AD.Name = AN.Name 
			AND 
			AD.Species = AN.Species
		INNER JOIN 
		Persons AS P
			ON P.Email = AD.Adopter_Email;
```

```sql
-- here we get the 100 rows we a looking for, inside the brackets are done fisrt, although they are not needed this is the 
-- better format
-- The logical order is called the Chiastic Order, you can force the processing order of multiple inner and outer joins
-- using this method 



SELECT	*
FROM	Animals AS AN
		LEFT OUTER JOIN 
			(
				Adoptions AS AD
				INNER JOIN 
				Persons AS P 
					ON	AD.Adopter_Email = P.Email
			)
			ON AD.Name = AN.Name 
			AND 
			AD.Species = AN.Species;
```

```sql
-- another example of chiastic order, notice the indentation, the brackets, alias used. When you write these start with the inner 
-- join first and then the outer join. 


SELECT	A.Name,
		A.Species,
		A.Breed,
		A.Primary_Color,
		V.Vaccination_Time,
		V.Vaccine,
		P.First_Name,
		P.Last_Name,
		SA.Role
FROM	Animals AS A
		LEFT OUTER JOIN
		(	Vaccinations AS V
			INNER JOIN
			Staff_Assignments AS SA
				ON SA.Email = V.Email
			INNER JOIN
			Persons AS P
				ON P.Email = V.Email
		)
		ON	A.Name = V.Name
			AND
			A.Species = V.Species
ORDER BY A.Species, A.Name, A.Breed, V.Vaccination_Time DESC;
```

As specified in the SQL standard, ternary logic, or three-valued logic (3VL), is a logic system with three truth values: TRUE, FALSE, and UNKNOWN. UNKNOWN is represented by NULL. Ternary logic applies to the evaluation of Boolean expressions, as well as predicates, and affects the results of logical operations such as AND, OR, and NOT:

- When used in expressions (e.g. [SELECT](https://docs.snowflake.com/en/sql-reference/sql/select.html) list), UNKNOWN results are returned as NULL values.
    
- When used as a predicate (e.g. [WHERE](https://docs.snowflake.com/en/sql-reference/constructs/where.html) clause), UNKNOWN results evaluate to FALSE.
    
- <span style="color: rgb(102, 102, 102); font-family: roboto, sans-serif; font-size: 18px;">Only records that evaluate to “true” in the WHERE clause are part of the query result set</span>  
    
- <span style="color: rgb(102, 102, 102); font-family: roboto, sans-serif; font-size: 18px;">1000 + Null = Null, therefore any expression with null = null !!&nbsp;</span>

```sql
-- Test_Employees Table contains many nulls. Lets see what we can do about this
select *
FROM dbo.test_employee
```

```sql
-- here the rows that contain null in any column are ignored as well as those less than or equal to 1500
-- where is Stephanie ???
SELECT * FROM test_employee  WHERE bonus + salary > 1500;
```

```sql
-- we can solve the above by using the below code, or we can add a new columns with the corrected amounts ideally in a view see 
-- NewC View     ISNULL(Salary, 0) AS Salary_Corrected
SELECT * FROM Test_Employee where isnull(salary, 0 ) + isnull(bonus, 0 ) > 1500
```

```sql
-- we can create a new column as another amonut, this should be done in a view, then you can use where statement....

SELECT *, ISNULL(Salary, 0) AS Salary_Corrected, ISNULL(Bonus, 0) AS Bonus_Corrected,  ISNULL(Salary, 0) + ISNULL(Bonus, 0) AS Total_Salary

FROM Test_Employee
```

```sql
-- taking the above example lets use a CTE (common table expression to filter the derived column Total Salary)
-- better to do this in a view...

WITH SALARY_FILTER (Employee_Name, Total_Salary)
AS
       (
            SELECT Employee_Name, ISNULL(Salary, 0) + ISNULL(Bonus, 0) AS Total_Salary

            FROM Test_Employee
        )
SELECT Employee_Name, Total_Salary
FROM SALARY_FILTER
WHERE Total_Salary > 1000



/*-- Define the CTE expression name and column list.  
WITH Sales_CTE (SalesPersonID, SalesOrderID, SalesYear)  
AS  
-- Define the CTE query.  
(  
    SELECT SalesPersonID, SalesOrderID, YEAR(OrderDate) AS SalesYear  
    FROM Sales.SalesOrderHeader  
    WHERE SalesPersonID IS NOT NULL  
)  
-- Define the outer query referencing the CTE name.  
SELECT SalesPersonID, COUNT(SalesOrderID) AS TotalSales, SalesYear  
FROM Sales_CTE  
GROUP BY SalesYear, SalesPersonID  
ORDER BY SalesPersonID, SalesYear;  
*/
```
